```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df= pd.read_csv ("especies3.csv", sep=";")
df.head(10)
```

```python
df.replace(['-'], pd.NA, inplace=True)
```

```python
df["Grupo"].value_counts(ascending=False).plot.bar(title= "Quantide de espécies ameaçadas por Grupo", color = "red");
plt.xlabel("Grupos")
plt.ylabel("Quantidade")
plt.legend()
plt.savefig("save.png")
```

```python
from pandas import DataFrame
CR = "CR"
df["Sigla Categoria de Ameaça"].value_counts(ascending=False).plot.barh(color="darkorange")
plt.bar(CR, height = 0.8, color = "darkorange")         
plt.xlabel("Quantidade")
plt.ylabel("Categoria de ameaça")

plt.legend();
```
